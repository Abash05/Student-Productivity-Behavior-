```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8)) 

sns.heatmap(df[['student_id','age','gender','study_hours_per_day', 'smartphone_usage_hours', 'social_media_hours', 'gaming_hours', 'sleep_hours', 'exercise_hours', 'class_attendance_percent', 'final_exam_score']].corr(), 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm',
            annot_kws={"size": 12}) 

plt.show()
```
